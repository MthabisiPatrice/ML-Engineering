```python
import json

with open("troubleshooting_knowledge_base.json", "r") as f:
    knowledge_base = json.load(f)

print(f"Loaded {len(knowledge_base)} issues.")
```

```python
#!pip install -U sentence-transformers
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
issue_sentences = []

for item in knowledge_base:
    issue_sentences.extend(item["keywords"])

issue_embeddings = model.encode(issue_sentences, convert_to_tensor=True)
```

```python
def find_issue(user_input):

    query_embedding = model.encode(user_input, convert_to_tensor=True)

    scores = util.cos_sim(query_embedding, issue_embeddings)

    best_score = scores.max().item()

    best_index = scores.argmax().item()

    if best_score < 0.40:
        return None

    count = 0

    for issue in knowledge_base:
        for keyword in issue["keywords"]:
            if count == best_index:
                return issue
            count += 1
```

```python
def troubleshoot():

    print("="*50)
    print(" AI Troubleshooting Assistant")
    print("="*50)

    problem = input("\nDescribe your issue:\n")

    issue = find_issue(problem)

    if issue is None:
        print("\nSorry, I couldn't identify the issue.")
        return

    print(f"\nPossible Issue: {issue['issue'].title()}")

    print("\nLet's diagnose it.\n")

    for question in issue["questions"]:

        answer = input(question + " (yes/no): ")

        if answer.lower() == "no":
            print("\nSuggested Fix:")

            print(random.choice(issue["solution"]))

            return

    print("\nRecommended Solutions")

    for solution in issue["solution"]:
        print("•", solution)
```

```python
def automate_fix(category):

    print("\nAttempting automatic repair...")

    if category == "network":

        print("✓ Resetting network adapter...")

        print("✓ Renewing IP address...")

        print("✓ Flushing DNS cache...")

        print("Done!")

    elif category == "software":

        print("✓ Checking updates...")

        print("✓ Clearing application cache...")

    else:

        print("Automation unavailable.")
```

```python
def chat_agent():

    print("\n🤖 Troubleshooting Agent Online")
    print("Type 'exit' to stop\n")

    while True:

        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Agent: Goodbye!")
            break

        issue = find_issue(user_input)

        if issue is None:
            print("Agent: Sorry, I couldn't identify the issue. Can you rephrase?")
            continue

        print(f"\nAgent: I think this is related to → {issue['issue']}\n")

        for q in issue["questions"]:
            answer = input(f"{q} (yes/no): ")

            if answer.lower() == "no":
                print("\nAgent: Suggested Fix:")
                print(random.choice(issue["solution"]))
                break

        automate_fix(issue["category"])
        print("\nAgent: Anything else?\n")
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from sentence_transformers import util


def show_chat_ui():
    problem_input = widgets.Text(
        value="",
        placeholder="Describe your issue",
        description="Problem:",
        layout=widgets.Layout(width="400px")
    )
    diagnose_button = widgets.Button(description="Diagnose")
    output_area = widgets.Output()

    def on_diagnose(_):
        with output_area:
            clear_output(wait=True)
            problem = problem_input.value.strip()

            if not problem:
                print("Please enter a problem first.")
                return

            issue = find_issue(problem)

            if issue is None:
                print("Sorry, I couldn't identify the issue. Try a different description.")
                return

            print(f"Possible issue: {issue['issue']}")
            print("\nSuggested steps:")
            for step in issue["solution"]:
                print("•", step)

            automate_fix(issue["category"])

    diagnose_button.on_click(on_diagnose)

    display(widgets.VBox([
        widgets.HTML("<b>AI Troubleshooting Assistant</b>"),
        problem_input,
        diagnose_button,
        output_area
    ]))


show_chat_ui()
```
