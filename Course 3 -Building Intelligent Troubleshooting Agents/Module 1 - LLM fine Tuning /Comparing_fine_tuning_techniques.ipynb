```python
# ============================================================
# fine_tuning_comparison.ipynb
# Dataset: IMDb sentiment (subset) via HuggingFace datasets
# ============================================================

# ── CELL 1: Install dependencies ──────────────────────────
#!pip install torch torchvision --force-reinstall --index-url https://download.pytorch.org/whl/cpu -q
#!pip install --upgrade transformers datasets peft accelerate bitsandbytes huggingface_hub -q

# ── CELL 2: Imports & Setup ───────────────────────────────
import time, torch
import numpy as np
from datasets import load_dataset
from transformers import (
    BertTokenizer, BertForSequenceClassification,
    Trainer, TrainingArguments, DataCollatorWithPadding
)
from peft import get_peft_model, LoraConfig, TaskType
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {DEVICE}")


# ── CELL 3: Load & Subset Dataset ────────────────────────
from datasets import load_dataset

# Fix: newer HuggingFace requires namespace/name format
raw = load_dataset("stanfordnlp/imdb")

# Use a small subset so Colab doesn't time out
train_subset = raw["train"].shuffle(seed=42).select(range(1600))
val_subset   = raw["test"].shuffle(seed=42).select(range(200))
test_subset  = raw["test"].shuffle(seed=42).select(range(200, 400))

print(f"Train: {len(train_subset)}, Val: {len(val_subset)}, Test: {len(test_subset)}")

# ── CELL 4: Tokenize ─────────────────────────────────────
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(batch["text"], truncation=True, max_length=256)

train_tok = train_subset.map(tokenize, batched=True)
val_tok   = val_subset.map(tokenize, batched=True)
test_tok  = test_subset.map(tokenize, batched=True)

# HuggingFace Trainer expects "label" not "labels"
for ds in [train_tok, val_tok, test_tok]:
    ds.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

collator = DataCollatorWithPadding(tokenizer)

# ── CELL 5: Metrics helper ────────────────────────────────
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    p, r, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    return {"accuracy": acc, "precision": p, "recall": r, "f1": f1}

# ── CELL 6: Training args factory ────────────────────────
def get_args(output_dir, epochs=2):
    return TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=epochs,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=16,
        eval_strategy="epoch",      # Changed from evaluation_strategy
        save_strategy="no",
        logging_steps=50,
        fp16=(DEVICE == "cuda"),
        report_to="none",
    )

results = {}
```

```python
from transformers import BertTokenizer

# Fix common dataset / environment issues before training
import datasets.config as ds_config
ds_config.TORCHVISION_AVAILABLE = False


if "tokenizer" not in globals():
    tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

raw = None
for name in ["stanfordnlp/imdb", "imdb"]:
    try:
        raw = load_dataset(name)
        print(f"Loaded dataset: {name}")
        break
    except Exception as exc:
        print(f"Failed to load {name}: {exc}")

if raw is None:
    raise RuntimeError("Could not load the IMDb dataset. Check the dataset name and network access.")

train_subset = raw["train"].shuffle(seed=42).select(range(1600))
val_subset = raw["test"].shuffle(seed=42).select(range(200))
test_subset = raw["test"].shuffle(seed=42).select(range(200, 400))

print(f"Train: {len(train_subset)}, Val: {len(val_subset)}, Test: {len(test_subset)}")

def tokenize(batch):
    return tokenizer(batch["text"], truncation=True, max_length=256)

train_tok = train_subset.map(tokenize, batched=True)
val_tok = val_subset.map(tokenize, batched=True)
test_tok = test_subset.map(tokenize, batched=True)

for ds in [train_tok, val_tok, test_tok]:
    ds.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

collator = DataCollatorWithPadding(tokenizer)
```

```python
# ════════════════════════════════════════════════════════════
# STEP 2 — Traditional Fine-Tuning (all params updated)
# ════════════════════════════════════════════════════════════
print("\n=== TRADITIONAL FINE-TUNING ===")

model_trad = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2
)

trainer_trad = Trainer(
    model=model_trad,
    args=get_args("./results_traditional"),
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=collator,
    compute_metrics=compute_metrics,
)

t0 = time.time()
trainer_trad.train()
trad_time = time.time() - t0

trad_metrics = trainer_trad.evaluate(test_tok)
trad_params  = sum(p.numel() for p in model_trad.parameters())

results["Traditional"] = {
    "time_s": round(trad_time, 1),
    "trainable_params": trad_params,
    **trad_metrics
}
print(f"Done in {trad_time:.1f}s | Params: {trad_params:,}")
print(trad_metrics)
```

```python
# ════════════════════════════════════════════════════════════
# STEP 3 — LoRA Fine-Tuning
# ════════════════════════════════════════════════════════════
print("\n=== LoRA FINE-TUNING ===")

base_lora = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2
)

lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    r=8,                          # rank of update matrices
    lora_alpha=16,                # scaling factor
    target_modules=["query", "value"],   # attention layers only
    lora_dropout=0.1,
    bias="none",
)

model_lora = get_peft_model(base_lora, lora_config)
model_lora.print_trainable_parameters()

trainer_lora = Trainer(
    model=model_lora,
    args=get_args("./results_lora"),
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=collator,
    compute_metrics=compute_metrics,
)

t0 = time.time()
trainer_lora.train()
lora_time = time.time() - t0

lora_metrics = trainer_lora.evaluate(test_tok)
lora_trainable = sum(p.numel() for p in model_lora.parameters() if p.requires_grad)

results["LoRA"] = {
    "time_s": round(lora_time, 1),
    "trainable_params": lora_trainable,
    **lora_metrics
}
print(f"Done in {lora_time:.1f}s | Trainable params: {lora_trainable:,}")
print(lora_metrics)
```

```python
#THis code wont run because the CUDA device is not available in this environment. You may need to run this code on a machine with a compatible GPU and CUDA installed.
# ════════════════════════════════════════════════════════════
# STEP 4 — QLoRA Fine-Tuning (4-bit quantization + LoRA)
# ════════════════════════════════════════════════════════════
print("\n=== QLoRA FINE-TUNING ===")

from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)

# Note: BitsAndBytes 4-bit works best on GPU; falls back gracefully on CPU
try:
    base_qlora = BertForSequenceClassification.from_pretrained(
        "bert-base-uncased",
        num_labels=2,
        quantization_config=bnb_config,
        device_map="auto",
    )
    qlora_config = LoraConfig(
        task_type=TaskType.SEQ_CLS,
        r=8,
        lora_alpha=16,
        target_modules=["query", "value"],
        lora_dropout=0.1,
        bias="none",
    )
    model_qlora = get_peft_model(base_qlora, qlora_config)
    model_qlora.print_trainable_parameters()
except Exception as e:
    print(f"4-bit load failed ({e}), falling back to 8-bit load_in_8bit")
    base_qlora = BertForSequenceClassification.from_pretrained(
        "bert-base-uncased", num_labels=2, load_in_8bit=True, device_map="auto"
    )
    model_qlora = get_peft_model(base_qlora, qlora_config)

qlora_args = get_args("./results_qlora")
# fp16 conflicts with 4-bit; disable
qlora_args.fp16 = False

trainer_qlora = Trainer(
    model=model_qlora,
    args=qlora_args,
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=collator,
    compute_metrics=compute_metrics,
)

t0 = time.time()
trainer_qlora.train()
qlora_time = time.time() - t0

qlora_metrics = trainer_qlora.evaluate(test_tok)
qlora_trainable = sum(p.numel() for p in model_qlora.parameters() if p.requires_grad)

results["QLoRA"] = {
    "time_s": round(qlora_time, 1),
    "trainable_params": qlora_trainable,
    **qlora_metrics
}
print(f"Done in {qlora_time:.1f}s | Trainable params: {qlora_trainable:,}")
print(qlora_metrics)
```

```python








# ════════════════════════════════════════════════════════════
# STEP 5 — Compare Results
# ════════════════════════════════════════════════════════════
import pandas as pd

summary = []
for method, r in results.items():
    summary.append({
        "Method":            method,
        "Train Time (s)":    r.get("time_s"),
        "Trainable Params":  f"{r.get('trainable_params', 0):,}",
        "Accuracy":          round(r.get("eval_accuracy", 0), 4),
        "F1":                round(r.get("eval_f1", 0), 4),
        "Precision":         round(r.get("eval_precision", 0), 4),
        "Recall":            round(r.get("eval_recall", 0), 4),
    })

df = pd.DataFrame(summary)
print("\n=== COMPARISON TABLE ===")
print(df.to_string(index=False))
```
